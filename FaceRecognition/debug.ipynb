1. 冻结模型参数

```python
import torch
from models.iresnet import iresnet50, iresnet100
```

```python
model = iresnet50(10177)
```

```python
for name, parameter in model.named_parameters():
    # print(name)
    if name in ["weight", "fc.weight", "fc.bias"]:
        parameter.requries_grad = True
    else:
        parameter.requries_grad = False
```

```python
model.conv1.weight
```

```python

for name,param in model.state_dict(keep_vars=True).items():
    if name in ["weight", "fc.weight", "fc.bias"]:
        param.requires_grad = True
    else:
        param.requires_grad = False
    print(name,param.requires_grad)
```

```python
for name,param in model.state_dict(keep_vars=True).items():
    print(name,param.requires_grad)
```

```python
filter(lambda p: p.requires_grad, model.parameters())
```

```python
for k,v in model.named_parameters():
     if k not in ["weight", "fc.weight", "fc.bias"]:
         v.requires_grad=False#固定参数
```

```python
for k,v in model.named_parameters():
    if k not in ["weight", "fc.weight", "fc.bias"]:
        print(v.requires_grad)#理想状态下，所有值都是False
```

```python
for name, param in model.named_parameters():
    if param.requires_grad:
        print(name)
```
